```python
# Demo Linear Regression

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# 1. Đọc dữ liệu
df = pd.read_csv("C:\\Users\\ASUS\\Desktop\\LinearRegression\\Data\\height_weight.csv")
X = df[["Height(cm)"]]
y = df["Weight(kg)"]

# 2. Huấn luyện với sklearn
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

plt.scatter(X, y, color="red", label="Thực tế")
plt.plot(X, y_pred, color="blue", label="Dự đoán (sklearn)")
plt.legend()
plt.show()

# 3. So sánh với numpy
X_np = np.array(df["Height(cm)"]).reshape(-1, 1)
y_np = np.array(df["Weight(kg)"])

ones = np.ones_like(X_np)
Xbar = np.concatenate((ones, X_np), axis=1)
w = np.linalg.pinv(Xbar.T @ Xbar) @ (Xbar.T @ y_np)

plt.scatter(X_np, y_np, color="red", label="Thực tế")
plt.plot(X_np, X_np * w[1] + w[0], color="green", label="Dự đoán (numpy)")
plt.legend()
plt.show()
```
